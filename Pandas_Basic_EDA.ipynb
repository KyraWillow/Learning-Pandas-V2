## Data Cleaning

Menangani missing value

```python
df.isna().sum() / len(df) * 100
```

```python
# mengisi datanya -> Unknown -> .fillna()
df["country"] = df["country"].fillna("Unknown")
```

```python
df["country"].isna().sum()
```

```python
df["children"] = df["children"].fillna(0)
```

```python
df["children"].isna().sum()
```

```python
df["agent"] = df["agent"].fillna(0)
```

```python
df["agent"].isna().sum()
```

```python
df = df.drop("company", axis=1)
```

```python
df.isna().sum()
```

```python
# duplicated dataabs
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
df.duplicated().sum()
```

## Aggregation & Features Engineering

```python
# .gorupby()
# KitA mencari sebuah Insight dalam data, melihat berapa total booking di hotel yang tidak/dicancele
df.groupby("hotel")["is_canceled"].value_counts().unstack()
```

```python
# Melihat rata rata harga untuk setiap hotel (membandingkan 2 buah fitur)
df.groupby("hotel")["adr"].mean()
```

```python
# Melihat harga rata rata pertahun untuk setiap hotel
df.groupby(["arrival_date_year", "hotel"])["adr"].mean().unstack()
```

```python
# Menggunakan aggregation lebih dari 1
df.groupby("customer_type")["lead_time"].agg(["min", "max", "mean", "count"])
```

### Analisis Pendapatan (ADR) dan Permintaan Pasar
Analisis ini membantu memahami segmen pasar mana yang paling menguntungkan dan bagaimana tren harga sepanjang tahun.

- Tujuan: Melihat bagaimana rata-rata tarif harian (ADR - Average Daily Rate) berfluktuasi berdasarkan segmen pasar dan bulan kedatangan.

```python
df.groupby(["market_segment", "arrival_date_month"])["adr"].mean().unstack()
```

## Features Engineering
Membuat fitur baru berdasarkan fitur yang sudah ada

```python
# Membuat fitur baru untuk total stay nights
df["total_nights"] = df["stays_in_weekend_nights"] + df["stays_in_week_nights"]
df[["total_nights", "stays_in_weekend_nights", "stays_in_week_nights"]].sample(10)
```

```python
# Menghandle tipe data date and time
df["reservation_status_date"] = pd.to_datetime(df["reservation_status_date"])
```

```python
df["reservation_status_date"].dtypes
```

```python
# .map() hanya bekerja pada Series (1 kolom).
month_mapping = {
    'January': 1,
    'February': 2,
    'March': 3,
    'April': 4,
    'May': 5,
    'June': 6,
    'July': 7,
    'August': 8,
    'September': 9,
    'October': 10,
    'November': 11,
    'December': 12}
df["arrival_date_month"] = df["arrival_date_month"].map(month_mapping)
```
