## Data Cleaning

Menangani missing value

```python
df.isna().sum() / len(df) * 100
```

```python
# mengisi datanya -> Unknown -> .fillna()
df["country"] = df["country"].fillna("Unknown")
```

```python
df["country"].isna().sum()
```

```python
df["children"] = df["children"].fillna(0)
```

```python
df["children"].isna().sum()
```

```python
df["agent"] = df["agent"].fillna(0)
```

```python
df["agent"].isna().sum()
```

```python
df = df.drop("company", axis=1)
```

```python
df.isna().sum()
```

```python
# duplicated dataabs
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
df.duplicated().sum()
```
